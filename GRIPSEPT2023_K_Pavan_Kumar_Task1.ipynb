### Cluster Analysis

#### Elbow Method

```python
iris_num=iris.iloc[:,0:4]
iris_num
```

```python
l=[]
for i in range(1,10):
    kmeans=KMeans(n_clusters=i,max_iter=50)
    kmeans.fit(iris_num)
    l.append(kmeans.inertia_)
```

```python
plt.plot(range(1,10),l)
plt.xlabel('Number Of Clusters')
plt.ylabel('Within Cluster Sum of Squares')
plt.title('Finding Optimal Number of Clusters through Elbow Method')
```

The optimal number of clusters is 3.

```python
kmeans=KMeans(n_clusters=3,max_iter=50)
iris_predict=kmeans.fit_predict(iris_num)
```

```python
iris_predict
```

```python
plt.scatter(iris_num.iloc[:,0],iris_num.iloc[:,1],c=iris_predict,cmap='viridis')
plt.title('Iris Species Clustered using K-Means')
plt.show()
sns.scatterplot(x='SepalLengthCm',y='SepalWidthCm',hue='Species',data=iris)
plt.title('Iris Species Actual Cluster')
plt.show()
```

Thus, we can say that cluster id 0 corresponds to Iris Species 'Versicolor', 1 corresponds to 'Setosa', 2 corresponds to 'Virginica'

```python
iris[iris_predict==2].Species.value_counts()
```

```python
plt.scatter(iris_num.iloc[iris_predict==0,0],iris_num.iloc[iris_predict==0,1],c='red',label='Versicolor')
plt.scatter(iris_num.iloc[iris_predict==1,0],iris_num.iloc[iris_predict==1,1],c='blue',label='Setosa')
plt.scatter(iris_num.iloc[iris_predict==2,0],iris_num.iloc[iris_predict==2,1],c='green',label='Virginica')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c='orange',label='Centroid')
plt.title('Iris Species Clustered using K-Means')
plt.legend(loc='best')
plt.show()
```
